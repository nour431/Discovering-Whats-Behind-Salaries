**Handling missing data**
>notice:
> 
> values of (BasePay,OvertimePay,OtherPay,TotalPay,TotalPayBenefits) sometimes can be negative values so im assuming that this is a fine <br>
>
> TotalPay and TotalPayBenefits sometimes = 0 or negative values too i will ignore that<br>
>
> <span style="color:red;">informations like this depends on the bussiness sometimes its fine to have a negative values because of (Deductions or Adjustments  or Reimbursements or Paybacks )so in this situations it will be better go get better understanding to be able to make a decision</span>
>


- you can use imputer to handle the missing values but in this case its not suitable because you can calculate the missing values of one column using the other columns 

- you can use `fillna()` also

- probelm :

  you have alot of missing values in the basepay and benefits columns so it will not be efficient to just drop it since there's no enough data.<br>


- My approach : 

  - Basepay column

    i can get the values of the basepay and benefits also using other columns 

    - assuming that :

      TotalPay = BasePay+OvertimePay+OvertimePay    

    - then i should be able to calculate the basepay    

      BasePay = TotalPay - OvertimePay - OvertimePay 

  - benefits column

    - assuming that :

      TotalPayBenefits= TotalPay+Benefits  

    - then i should be able to calculate the basepay    

      Benefits = TotalPayBenefits - TotalPay 
  

```python
df =df.drop(['Notes','Status'] ,axis=1)
```

```python

# Calculate missing BasePay values
df.loc[df['BasePay'].isnull(),'BasePay'] = df['TotalPay'] - df['OvertimePay'] - df['OtherPay']
# Calculate missing Benefits values
df.loc[df['Benefits'].isnull(),'Benefits'] = df['TotalPayBenefits'] - df['TotalPay']
```

```python
df.isna().sum()
```

```python
df.dropna(subset=['BasePay','OvertimePay' ,'OtherPay'],inplace=True)
df.isna().sum()
```
